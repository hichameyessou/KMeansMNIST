### KMEANS: Visualization of centroids on MNIST dataset 

```python
import pandas as pd
import numpy as np
import sys
import re
import matplotlib.pyplot as plt
images = []
tmp = []

with open(r'C:\Users\Hichame\Documents\KMeansMNIST\target\clusters', 'r') as f:
    for line in f:
        tmp = []
        line = re.sub('[\[\]]', '', line)
        for item in line.split(','):
            tmp.append(float(item))
        images.append(tmp)
```

```python
def drawDigit(digit):
    image = np.array(digit)
    plt.figure()
    digit = plt.imshow(image.reshape(28,28))
    digit.set_cmap('gray_r')
    plt.show()
```

```python
for index in images:
    drawDigit(index)
```
